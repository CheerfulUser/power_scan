```python
prf = TESS_PRF(4,4,20,1000,1000)
```

```python
target1 = prf.locate(7.2,7.1,stampsize=(25,25))
target2 = prf.locate(8.8,7.6,stampsize=(25,25))
target3 = prf.locate(20.3,20.4,stampsize=(25,25))

frame = np.arange(1,10000) * 1.0

signal = 5*np.sin(1/1000*frame*2*np.pi)
target1 = target1[np.newaxis,:,:] * signal[:,np.newaxis,np.newaxis]

signal = 10*np.sin(1/23.2*frame*2*np.pi)
target2 = target2[np.newaxis,:,:] * signal[:,np.newaxis,np.newaxis]

signal = 100*np.sin(1/2000*frame*2*np.pi)
target3 = target3[np.newaxis,:,:] * signal[:,np.newaxis,np.newaxis]

noise = np.random.normal(loc=0,scale=2,size=(target1.shape))

scene = target1+target2+target3+noise

```
